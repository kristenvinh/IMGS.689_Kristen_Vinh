```python
import matplotlib.pyplot as plt
import numpy as np

# Define the input filename
filename = 'vegetation.tree.quercus.agrifolia.vswir.jpl108.jpl.asd.spectrum.txt'

try:
    # Load the data from the text file.
    # The header has 24 lines, so we skip them.
    data = np.loadtxt(filename, skiprows=24)

    # Separate the data into wavelength (X) and reflectance (Y)
    wavelength = data[:, 0]
    reflectance = data[:, 1]/100

    # Create the plot
    plt.style.use('seaborn-v0_8-whitegrid') # Use a nice style for the plot
    plt.figure(figsize=(10, 6)) # Set the figure size for better readability

    plt.plot(wavelength, reflectance, label='Quercus agrifolia', color='green')

    # Add labels and a title
    plt.title('Spectral Reflectance of Quercus agrifolia (Coast Live Oak) [cite: 1]')
    plt.xlabel('Wavelength (micrometers) ')
    plt.ylabel('Reflectance (percentage) ')
    plt.legend()
    plt.grid(True) # Add a grid for easier reading

    # Display the plot
    plt.show()

except FileNotFoundError:
    print(f"Error: The file '{filename}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the input filename
filename = 'manmade.road.pavingasphalt.solid.all.0674uuuasp.jhu.becknic.spectrum.txt'

try:
    # Load the data from the text file.
    # The header has 22 lines, so we skip them.
    data = np.loadtxt(filename, skiprows=22)

    # Separate the data into wavelength (X) and reflectance (Y)
    wavelength = data[:, 0]
    reflectance = data[:, 1]/100

    # --- Filter the data to the desired wavelength range ---
    # Create a boolean mask for wavelengths between 0.35 and 2.5 µm
    mask = (wavelength >= 0.35) & (wavelength <= 2.5)
    
    # Apply the mask to get the filtered data
    wavelength_filtered = wavelength[mask]
    reflectance_filtered = reflectance[mask]

    # Create the plot
    plt.style.use('seaborn-v0_8-whitegrid')
    plt.figure(figsize=(10, 6))

    # Plot the filtered data
    plt.plot(wavelength_filtered, reflectance_filtered, label='Construction Asphalt', color='black')

    # Add labels and a title based on the file's header
    plt.title('Spectral Reflectance of Construction Asphalt')
    plt.xlabel('Wavelength (micrometers)')
    plt.ylabel('Reflectance (percent)')
    
    # Set the x-axis limits to the requested range
    plt.xlim(0.35, 2.5)
    
    plt.legend()
    plt.grid(True)

    # Display the plot
    plt.show()

except FileNotFoundError:
    print(f"Error: The file '{filename}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```
